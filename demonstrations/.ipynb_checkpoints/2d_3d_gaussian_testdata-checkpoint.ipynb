# 3D Gaussian

```python
grid = np.zeros([100,100,100])
amplitude = 50
mu_x, sigma_x = 50, 5
mu_y, sigma_y = 45, 5
mu_z, sigma_z = 30, 5

n_neighbors_x = int(3 * sigma_x ) + 1
n_neighbors_y = int(3 * sigma_y ) + 1
n_neighbors_z = int(3 * sigma_z ) + 1

for kx in range(max(0, mu_x - n_neighbors_x), min(grid.shape[0], mu_x + n_neighbors_x)):
    for ky in range(max(0, mu_y - n_neighbors_y), min(grid.shape[1], mu_y + n_neighbors_y)):
        for kz in range(max(0, mu_z - n_neighbors_z), min(grid.shape[1], mu_z + n_neighbors_z)):
#                for ky in range(max(0, iy - n_neighbors), min(ngrid, iy + n_neighbors)):
            d2 = (kx-mu_x)**2 + (ky-mu_y)**2+ (kz-mu_z)**2
            value = amplitude*np.exp(-d2 / (2*sigma_x*sigma_y*sigma_z))
            grid[kx, ky, kz] += value
```
